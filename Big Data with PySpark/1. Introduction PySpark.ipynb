PySpark sebagai alat komputasi yang dapat mengakses big data secara cepat mempunyai fitur sebagai berikut :


<ul>
  <li><p align="justify">Pemrosesan dalam memori (<i>in-memory computation</i>), merujuk pada pendekatan pemrosesan data di mana data disimpan dan diproses di dalam memori komputer, daripada di penyimpanan disk yang lebih lambat. </p></li>
  <li><p align="justify"><i>Distributed processing</i> menggunakan parallelize</p></li>
  <li><p align="justify">Dapat digunakan pada beberapa cluster managers (<i>Spark, Yarn,</i> dll)</p></li>
  <li><p align="justify">Kemampuan sistem untuk tetap berfungsi dengan baik dan memberikan output yang benar meskipun terjadi kegagalan (<i>Fault Tolerant</i>)</p></li>
  <li><p align="justify">Menerapkan strategi evaluasi di mana ekspresi atau operasi tidak dievaluasi atau dieksekusi secara segera, tetapi ditunda hingga benar-benar diperlukan (<i>Lazy Evaluation</i>)</p></li>
  <li><p align="justify">Menggunakan mekanisme penyimpanan sementara yang digunakan untuk menyimpan data yang sering diakses atau hasil komputasi agar dapat diakses dengan cepat. (<i>Cache & Persistence</i>)</p></li>
</ul></p>

<a id="4.1"></a>
## <b><span style='color:#211951'>4.1</span> <span style='color:#FF8911'>Membuat RDD</span></b> 

Eksekusi kode <code>spark.sparkContext.parallelize(data)</code> maka tipe data dasar Python (<i>iterables</i>) akan diubah menjadi RDD

<br>
<p align="right" style="font-size:80%;">
docs : <a href="https://spark.apache.org/docs/latest/api/python/reference/api/pyspark.SparkContext.parallelize.html"><i>https://spark.apache.org/docs/latest/api/python/reference/api/pyspark.SparkContext.parallelize.html</i>
</p>

````python
# Buat sebuah list biasa (list pada Python)
dataList = [4, 5, 2, 8, 1]

# Konversi menjadi RDD dengan cara parallelize
rdd = sc.parallelize(dataList)

# Tampilkan hasilnya
print(rdd)
````

<a id="4.2"></a>
## <b><span style='color:#211951'>4.2</span> <span style='color:#FF8911'>Operasi RDD</span></b> 

Transformasi:
* <code>map(func)</code>: Menerapkan fungsi pada setiap elemen RDD dan mengembalikan RDD baru.
* <code>filter(func)</code>: Mengembalikan RDD baru yang hanya berisi elemen yang memenuhi predikat yang diberikan.
* <code>flatMap(func)</code>: Mirip dengan map, tetapi setiap elemen input dapat dipetakan menjadi nol atau lebih elemen output.
* <code>reduceByKey(func)</code>: Menerapkan fungsi reduksi pada nilai-nilai setiap kunci, menghasilkan RDD baru berupa pasangan kunci-nilai.
* <code>sortByKey()</code>: Mengurutkan RDD berdasarkan kunci secara menaik.
* <code>join(other)</code>: Melakukan join dalam (inner join) antara dua RDD berdasarkan kunci yang sama.

Actions :
* <code>count()</code>: Mengembalikan jumlah total elemen dalam RDD.
* <code>collect()</code>: Mengembalikan semua elemen RDD sebagai array.
* <code>take(n)</code>: Mengembalikan n elemen pertama dari RDD.
* <code>foreach(func)</code>: Menerapkan fungsi pada setiap elemen RDD (berguna untuk mencetak atau menyimpan data).

<br>
<p align="right" style="font-size:80%;">
docs : <a href="https://spark.apache.org/docs/latest/rdd-programming-guide.html#rdd-operations"><i>https://spark.apache.org/docs/latest/rdd-programming-guide.html#rdd-operations</i>
</p>

<p align="left" style="font-size:120%;"><b>Contoh :</b></p>

````python
# Buat sebuah list biasa (list pada Python)
dataList = [4, 5, 2, 8, 1]

# Konversi menjadi RDD dengan cara parallelize
rdd = sc.parallelize(dataList)

# Hitung nilai maksimum
max_rdd = rdd.max()

# Tampilkan hasil
display(f'Nilai maksimum : {max_rdd}')
````

## <b><span style='color:#211951'>5.1</span> <span style='color:#FF8911'>Tipe Data</span></b> 


Tipe data pada dataframe spark adalah sebagai berikut

* `StringType`: Tipe data untuk representasi string.
* `IntegerType`: Tipe data untuk representasi bilangan bulat.
* `LongType`: Tipe data untuk representasi bilangan bulat panjang.
* `FloatType`: Tipe data untuk representasi bilangan pecahan (float).
* `DoubleType`: Tipe data untuk representasi bilangan pecahan (double).

* `BooleanType`: Tipe data untuk representasi nilai boolean (True/False).
* `DateType`: Tipe data untuk representasi tanggal.
* `TimestampType`: Tipe data untuk representasi tanggal dan waktu.
* `ArrayType`: Tipe data untuk representasi array.
* `MapType`: Tipe data untuk representasi map (kunci-nilai).
* `StructType`: Tipe data untuk representasi struktur yang terdiri dari beberapa kolom.<br><br>

<a id="5.2"></a>
## <b><span style='color:#211951'>5.2</span> <span style='color:#FF8911'>Membuat DataFrame</span></b> 


Untuk membuat DataFrame perlu mengeksekusi sintaks berikut
```
spark.createDataFrame(data, schema)
```
dengan data adalah data yang ingin dibuat
schema adalah nama kolom atau Struktur Kolom (Struct Field)<br><br>

<p align="right" style="font-size:80%;">
docs : <a href="https://spark.apache.org/docs/3.1.1/api/python/reference/api/pyspark.sql.SparkSession.createDataFrame.html"><i>https://spark.apache.org/docs/3.1.1/api/python/reference/api/pyspark.sql.SparkSession.createDataFrame.html</i>
</p>



<p align="left" style="font-size:120%;"><b>Contoh :</b> Membuat DataFrame Kosong</p>

````python
# Import library untuk tipe data
from pyspark.sql.types import *

# Inisialisasikan Nama Kolom - Tipe Data - dan nullable
schema = StructType([
    StructField('Nama', StringType(), nullable = True),
    StructField('Umur', IntegerType(), nullable = True)
])

# Buat DataFrame kosong
data_kosong = spark.createDataFrame(
    data = [],
    schema = schema
)

# Tampilkan hasilnya
data_kosong.show()
````

<p align="left" style="font-size:120%;"><b>Contoh :</b> Membuat DataFrame dari List Data</p>

````python
# Inisialisasi data
nasabah = [
    ('Agus', '1994-10-09', 'Priority', 500000000),
    ('Bambang', '1987-02-14', 'Silver', 5000000),
    ('Caknan', '1997-08-17', 'Gold', 8500000),
]

# Definisikan tipe data
schema = StructType([
    StructField('Nama', StringType(), nullable = True),
    StructField('Tanggal Lahir', DateType(), nullable = True),
    StructField('Tipe Nasabah', StringType(), nullable = True),
    StructField('Salary', DoubleType(), nullable = True),
])

# Buat DataFrame
data_nasabah = spark.createDataFrame(
    data = nasabah,
    schema = kolom
)
````

## <b><span style='color:#211951'>5.9</span> <span style='color:#FF8911'>Spark SQL</span></b> 

<p align="justify">
Spark SQL adalah modul dari Apache Spark yang menyediakan antarmuka pemrograman yang ditingkatkan untuk bekerja dengan data terstruktur, seperti data dalam format tabel, Hive, dan parquet. Spark SQL mengintegrasikan pemrosesan SQL tradisional dengan kemampuan pemrosesan data yang kuat dari Apache Spark.

Dengan Spark SQL, berbagai tugas terkait SQL pada data terstruktur, seperti menjalankan query SQL, melakukan agregasi, penggabungan data, pemrosesan kolom, dan pemodelan data. Spark SQL juga mendukung penulisan dan eksekusi UDF (User-Defined Functions) yang memungkinkan  kode ditulis khusus untuk memanipulasi data dalam bahasa pemrograman.
</p>

Jalankan sintaks
```
spark.sql(query)
```
untuk mengakses data dengan query<br><br>

<p align="right" style="font-size:80%;">
docs : <a href="https://spark.apache.org/sql/"><i>https://spark.apache.org/sql/</i>
</p>

<p align="left" style="font-size:120%;"><b>Contoh :</b></p>

````python
# Dari data_credit sebelumnya dibentuk Temporary View
data_loan.createOrReplaceTempView('temp_credit')
````

````python
# Proses querying
query = '''
  SELECT
      id,
      member_id,
      loan_amnt,
      funded_amnt
  FROM temp_credit
  WHERE loan_amnt > 10000
  ORDER BY member_id
'''

# Eksekusi query
data_loan_from_query = spark.sql(query)

# Tampilkan hasilnya
data_loan_from_query.show()
````
